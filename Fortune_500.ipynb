# FORTUNE 500 COMPANIES - 2020 : Insights from Data

The Fortune 500 is an annual list compiled and published by Fortune magazine that ranks 500 of the largest United States corporations by the annual total revenue using publicly available data.

The Fortune 500, created by Edgar P. Smith, was first published in 1955.





```python
import numpy as np 
import pandas as pd 
```

### Datatype Conversion : From object to float


```python
df = df.replace('-','0')
```

```python
def remove_comma(x):
  x = x.replace(',','')
  return x
```

```python
cols_comma_remove = ['no_of_employees','revenues','profits','assets','market_value' ]
for i in cols_comma_remove:
  df[i] = df[i].apply(remove_comma)
```

```python
for i in df.columns[2:]:
  df[i] = df[i].astype(float)
for i in df.columns:
  print(i, ' : ', df[i].dtype)
```

## Rank Analysis : 

```python
df['rank_is_changed'] = df['rank_change'].apply(lambda x : False if x==0.0 else True)
rank_change_df = df[df['rank_is_changed']== False]
rank_change_df.reset_index(drop=True, inplace=True)
rank_change_df.head()
```

```python
rank_increased_companies = df[df['rank_change']>0.0]
rank_increased_companies.reset_index(drop=True, inplace=True)
rank_increased_companies.head()
```

```python
rank_decreased_companies = df[df['rank_change']<0.0]
rank_decreased_companies.reset_index(drop=True, inplace=True)
rank_decreased_companies.head()
```

## Top 10 Companies of the Rank Analysis Table 

```python
no_change_list = rank_change_df['company'].tolist()
inc_list = rank_increased_companies['company'].to_list()
dec_list = rank_decreased_companies['company'].to_list()
top10_no_change = no_change_list[:10]
top10_inc = inc_list[:10]
top10_dec = dec_list[:10]
rank_data = {'Rank_Retained' : top10_no_change,
           'Rank_Increased' : top10_inc,
           'Rank_Decreased' : top10_dec
           }
rank_df = pd.DataFrame(rank_data)
rank_df[:10]
```

```python
rank_df.to_csv('Rank Analysis.csv')
```

```python
rank_analysis = [rank_change_df, rank_increased_companies, rank_decreased_companies]
for i in rank_analysis:
  print(len(i.index))
```
